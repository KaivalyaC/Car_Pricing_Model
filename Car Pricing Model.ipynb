# 4. Function Definations <a class="anchor" id="functions"></a>
    
Many times we might want to use same fuctionality multiple times in out code.<br> 
For example every time we manipulate our data we might want to check the central tendency of the data like mean median etc. Thats why we can create a function which can be called anytime we want the summary of the statistical data. It also give some idea about the outliers. If the std deviation of any feature is more than double the mean then it is considered that there is possibility of outliers<br> 
The second function is for displaying output of the model performance in terms of R2, Adj-R2, MAE, MAPE. Sklearn is very good with model creating and performance measurement but displaying thise results is little bit cumbersome. Thats why we are creating a separate function for it.
 
* [Go to Top](#index)

```python
def summaryStats(X,feature_list):
    temp = pd.DataFrame(dict(mean = X[feature_list].mean().round(2),
                             median = X[feature_list].median().round(2),
                             std_dev = X[feature_list].std().round(2),
                             min_val = X[feature_list].min().round(2),
                             max_val = X[feature_list].max().round(2),
                             outlier_issue_using_std = np.where(X[feature_list].std()/X[feature_list].mean() > 2,
                                                      True,False),
                             outlier_issue_using_median = np.where((X[feature_list].median()/X[feature_list].mean() > 1.25 ) | (X[feature_list].median()/X[feature_list].mean() < 0.75),
                                            True,False),
                             null_values = X[feature_list].isnull().sum()
                            ))
    return temp

def target_by_cats(X):
    '''
    This function assumes the input dataframe (X) has a numerically defined target
    variable in the first column and that all remaining features are categorical
    '''
    global_mean = X.iloc[:,0].mean()
    target = X.columns[0]
    for col in X.columns[1:]:
        df_group = X.groupby(by=col)[target].agg(['mean']).round(2)
        df_group['diff'] = (df_group['mean'] - global_mean).round(2)
        df_group['%diff'] = (df_group['diff']/df_group['mean'])*100
        display(df_group)

def trainTestMetrics(X_train,X_test,y_train,y_test,model):
    names=['R2','Adj-R2','MAE','MAPE','SSE'] 
    
    # Training Metrics
    y_hat = model.predict(X_train)
    
    # Create R2
    r2 = r2_score(y_train,y_hat).round(3)
    
    # Create adj_R2
    adj_r2 = (1-(1-r2_score(y_train,y_hat))*(len(y_train)-1)/(len(y_train)-len(model.coef_)-1)).round(3)

    
    train_metrics = [r2, adj_r2,
                     mean_absolute_error(y_train,y_hat).round(3),
                     mean_absolute_percentage_error(y_train,y_hat).round(3)*100,
                     (mean_squared_error(y_train,y_hat)*len(y_train)).round(3)
                    ]
    train_metrics = pd.DataFrame({'Train':train_metrics},index=names)
    
    # Testing Metrics
    y_hat = model.predict(X_test)
    test_metrics = [r2_score(y_test,y_hat).round(3),
                    (1-(1-r2_score(y_test,y_hat))*(len(y_test)-1)/(len(y_test)-len(model.coef_)-1)).round(3),
                    mean_absolute_error(y_test,y_hat).round(3),
                    mean_absolute_percentage_error(y_test,y_hat).round(3)*100,
                    (mean_squared_error(y_test,y_hat)*len(y_test)).round(3)
                   ]
    test_metrics = pd.DataFrame({'Test':test_metrics},index=names)
    
    all_metrics = train_metrics.merge(test_metrics,left_index=True,right_index=True)
    print(all_metrics)
```

```python
%%writefile app.py

import pickle
import streamlit as st
import math

pickle_in = open('classifier', 'rb')
classifier = pickle.load(pickle_in)

@st.cache()

# Define the function which will make the prediction using data
# inputs from users
def prediction(engine_hp, highway_mpg, year,
               make_type, vehicle_type, transmission_type):
    
    #initialization
    make_types = {'Economy':0,
           'Luxury':0,
           'Mid Level':0,
           'Ultra Luxury':0}
    
    vehicle_types = {'Convertible':0,
                    'Coupe':0,
                    'Hatchback':0,
                    'Pickup':0,
                    'Sedan':0,
                    'SUV':0,
                    'Van':0,
                    'Wagon':0}
    
    transmission_types = {'Automated Manual':0,
                         'Automatic':0,
                         'Direct Drive':0,
                         'Manual':0,
                         'Unknown':0}
    
    #value population
    for make in make_types:
        if make == make_type:
            make_types[make] = 1
            
    for style in vehicle_types:
        if style == vehicle_type:
            vehicle_types[style] = 1
            
    for transmission in transmission_types:
        if transmission == transmission_type:
            transmission_types[transmission] = 1
    
    
    # Make predictions
    prediction = classifier.predict(
        [[engine_hp,highway_mpg,year,
              make_types.get('Economy'),make_types.get('Luxury'),make_types.get('Mid Level'),make_types.get('Ultra Luxury'),
              vehicle_types.get('Convertible'),vehicle_types.get('Coupe'),
              vehicle_types.get('Hatchback'),vehicle_types.get('Pickup'),
              vehicle_types.get('Sedan'),vehicle_types.get('SUV'),
              vehicle_types.get('Van'),vehicle_types.get('Wagon'),
              transmission_types.get('Automated Manual'),transmission_types.get('Automatic'),
              transmission_types.get('Direct Drive'),transmission_types.get('Manual'),transmission_types.get('Unknown'),
              engine_hp*make_types.get('Economy'),engine_hp*make_types.get('Luxury'),
              engine_hp*make_types.get('Mid Level'),engine_hp*make_types.get('Ultra Luxury'),
              highway_mpg*transmission_types.get('Automated Manual'),highway_mpg*transmission_types.get('Automatic'),
              highway_mpg*transmission_types.get('Direct Drive'),highway_mpg*transmission_types.get('Manual'),highway_mpg*transmission_types.get('Unknown')
         ]])
    
    return 'Price for these features would be $' + str(round(math.exp(prediction),2))

# This is the main function in which we define our webpage
def main():
    
    # Create input fields
     
    engine_hp = st.number_input("What is the horse power of the engine?",
                                  min_value=0,
                                  max_value=600,
                                  value=210,
                                  step=30,
                                 )
    highway_mpg = st.number_input("What is the average miles per gallon on highway?",
                              min_value=0,
                              max_value=60,
                              value=30,
                              step=5
                             )

    year = st.number_input("What is the year of make?",
                              min_value=2000,
                              max_value=2020,
                              value=2020,
                              step=1
                             )
    make_type = st.selectbox('Select the type of the Make',
                             ('Economy', 'Mid Level', 'Luxury', 'Ultra Luxury'))
    
    vehicle_type = st.selectbox('Select the type of the Vehicle Style',
                             ('Convertible', 'Coupe', 'Hatchback', 'Pickup', 'Sedan', 'SUV', 'Van', 'Wagon'))
    
    transmission_type = st.selectbox('Select the type of the Transmission System',
                             ('Automatic','Manual','Automated Manual','Direct Drive','Unknown'))
    

    result = ""
    
    # When 'Predict' is clicked, make the prediction and store it
    if st.button("Predict the Price"):
        result = prediction(engine_hp, highway_mpg, year,
               make_type, vehicle_type, transmission_type)
        st.success(result)
        
if __name__=='__main__':
    main()
```
